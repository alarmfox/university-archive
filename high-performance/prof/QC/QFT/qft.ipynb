# Quantum Fourier Transform

```python
from qiskit import QuantumCircuit
import math
from qiskit.circuit.library import PhaseGate
```

## QFT circuit

```python
def my_qft(n):
    qc = QuantumCircuit(n)
    for i in range(n):
        qc.h(i)
        for j in range(i+1, n):
            rot = PhaseGate(2*math.pi/2**(j-i+1)).control(1)
            qc.append(rot, [j, i])
    for i in range(math.floor(n/2)):
        qc.swap(i, n-i-1)
    
    return qc
    
qc = my_qft(4)
qc.draw(output = 'mpl', style = 'clifford')
```

## QFT from Qiskit library

```python
from qiskit.circuit.library import QFT
qc1 = QFT(4)
qc1.decompose().draw(output = 'mpl', style = 'clifford')
```

## Show how the state changes
### $ \ket{0000} $ 

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

state = Statevector(qc)
state1 = Statevector(qc1)
plot_bloch_multivector(state, figsize=[3,3])
```

```python
plot_bloch_multivector(state1, figsize=[3,3])
```

### $ \ket{0001} $ 

```python
qc = QuantumCircuit(4)

qc.x(3)
qc.append(my_qft(4), range(4))

plot_bloch_multivector(Statevector(qc), figsize=[3,3])
```

```python
qc1 = QuantumCircuit(4)
qc1.x(0)
qc1.append(QFT(4), range(4))
plot_bloch_multivector(Statevector(qc1),  figsize=[3,3])
```
