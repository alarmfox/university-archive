# Soluzione di un problema di PL

Una azienda deve produrre 2 profilati metallici (A e B) che richiedono l'impiego di manodopera, disponibile al massimo in 36 squadre. Per la produzione giornaliera di un lotto di A e di un lotto di B si impiegano rispettivamente 3 e 6 squadre. È necessario produrre tra A e B almeno 4 lotti al giorno. Si possono impiegare due diverse tecnologie (1 e 2). La tecnologia 1 produce al massimo 2 lotti di B per ogni lotto di A (cioè le produzioni di B ed A sono al massimo nel rapporto 2:1). La tecnologia 2 produce al massimo 2 lotti di A per ogni lotto di B (cioè le produzioni di A e B sono al massimo nel rapporto 2:1). Il profitto unitario è lo stesso per entrambi i prodotti. Si vuole determinare il piano di produzione giornaliero che massimizza il profitto totale.

### Variabili

$x_{A}$ - numero di lotti di tipo A prodotti

$x_{B}$ - numero di lotti di tipo A prodotti


### Importa la libreria GRB inizializza il problema e definisce le variabili¶

```python
import gurobipy as gp
from gurobipy import GRB

# Inizializza il modelo
mod = gp.Model('esempio di PL')

# Crea le varibili di decisione

xA = mod.addVar(name="XA")
xB = mod.addVar(name="XB")
```

```python
xA
```

### Objective Function

- Massimizza il profitto (Il profitto unitario è lo stesso per entrambi  i prodotti)

\begin{equation}
\text{Maximize} \quad x_A + x_B 
\end{equation}

```python
obj = mod.setObjective(xA + xB, GRB.MAXIMIZE)
```

### Vincoli del problema

1) Per la produzione giornaliera di un lotto di A e di un lotto di B si impiegano rispettivamente 3 e 6 squadre. Si dispone di al massimo 36 squadre.

\begin{equation}
3 \ x_A + 6 \ x_B \leq 36
\end{equation}

```python
Vincolo_1 = mod.addConstr(3 * xA + 6 * xB <= 36, name="Primo_vincolo")
```

2) È necessario produrre tra A e B almeno 4 lotti al giorno.

\begin{equation}
x_A + x_B \geq 4
\end{equation}

```python
Vincolo_2 = mod.addConstr(xA + xB >= 4, name="Secondo_vincolo")
```

3) La tecnologia 1 produce al massimo 2 lotti di B per ogni lotto di A. 

\begin{equation}
- 2 \ x_A + x_B \leq 0
\end{equation}

```python
Vincolo_3 = mod.addConstr(- 2 * xA + xB <= 0, name="Terzo_vincolo")
```

4) La tecnologia 2 produce al massimo 2 lotti di A  per ogni lotto di B  

\begin{equation}
x_A - 2 \ x_B \leq 0
\end{equation}

```python
Vincolo_4 = mod.addConstr(xA - 2 * xB <= 0, name="Quarto_vincolo")
```

## Soluzione del problema

### Risolve il problema

```python
mod.optimize()
```

```python
if mod.status == GRB.OPTIMAL:

    # print solution

    print('\nObj: %g \n' % mod.objVal)
    
    print('xA = %g' % xA.x)
    print('xB = %g' % xB.x)

    #for x in mod.getVars():
    #    print('%s %g' % (x.varName, v.x))
```
