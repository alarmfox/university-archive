# Definizione di un grafo utilizzando la libreria Networkx

```python
import networkx as nx 
import matplotlib.pyplot as plt

# crea e stampa un grafo di 6 nodi e 3 archi
G0 = nx.DiGraph()
G0.add_edge(1, 2)
G0.add_edge(2, 3) 
G0.add_edge(2, 4) 
G0.add_node(5)
G0.add_node(6)

nx.draw_networkx(G0, with_labels = True, node_size=550)
```

### Legge i nodi del grafo da una lista

```python
# Costruisce un grago g di 9 nodi (i nodi sono etichettati da 1 a 9)
ListOfNodes = range(1,9)
G = nx.DiGraph()
G.add_nodes_from(ListOfNodes)
nx.draw_networkx(G, with_labels = True, node_size=550)
```

### E' possibile indicare la posizione di ogni nodo o di un sottoinsieme di nodi

```python
PosOfNodes ={1: (1, 10), 2: (7,10), 3: (10, 7), 4: (4,7), 5: (1,7), 6: (1,4), 7: (4,2), 8: (7,4), 9: (10,10)}
PosOfNodes
```

```python
plt.figure(figsize=(10,5))
nx.draw_networkx(G, PosOfNodes, with_labels=True, node_size=700)
plt.draw()
```

### Aggiunge gli archi leggendoli da una lista

```python
ListOfArcs = [(1,2), (1,5), (2,3), (2,4), (2,9), (3,4), (3,9), (4,6), (4,7), (4,8), (5,4), (5,6), 
              (7,6), (7,8), (8,2), (8,3)]
G.add_edges_from(ListOfArcs)
plt.figure(figsize=(10,5))
nx.draw_networkx(G, PosOfNodes, with_labels=True, node_size=700)
plt.draw()
```

### Il grafo può essere scritto e letto i divesri formati

https://networkx.org/documentation/stable/reference/readwrite/index.html


```python
nx.write_edgelist(G, "Grafo_G_edgelist.txt")
G1 = nx.read_edgelist("Grafo_G_edgelist.txt")
```

### Si possono associare dei pesi agli archi del grafo

```python
ListOfWeightedArcs = [(1,2,7.0), (1,5,9.0), (2,3,35.0), (2,4,15.0), (2,9,40.0), (3,4,21.0), (3,9,10.0), (4,6,28.0), 
                      (4,7,21.0), (4,8,11.0), (5,4,44.0), (5,6,69.0), (7,6,5.0), (7,8,7.0), (8,2,1.0), (8,3,1.0)]
G.add_weighted_edges_from(ListOfWeightedArcs)

weights = nx.get_edge_attributes(G,'weight')
weights
```

```python
plt.figure(figsize=(10,5))
nx.draw_networkx(G, PosOfNodes, with_labels=True, node_size=700)
nx.draw_networkx_edge_labels(G,PosOfNodes,edge_labels=weights,label_pos=0.75)
plt.draw()
```

```python
nx.write_edgelist(G, "Grafo_G_edgelist.txt")
```
