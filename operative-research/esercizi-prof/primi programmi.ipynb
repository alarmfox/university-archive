# Tuple

## Una tupla è una collezione ordinata, non modificabile, che può contenere valori duplicati

```python
# concatena tuple
t = (1, 2, 3)
t2 = 4, 5, 6
tutto = t + t2 + (8, 9) + ("dieci", "undici")
print(tutto)
print(tutto[3])
print(8 in tutto)
print(22 in tutto)
print("dieci" in tutto)
```

```python
# replica una tupla
t1 = (2, 3)
rep_t1 = t1 * 4
print(rep_t1)
```

```python
# conta il numero di volte che un elemento è presente in una tupla
l = (8, 9, "dieci", "undici", 9)
occorrenze_9 = l.count(9)
print("9 compare", occorrenze_9, "volte nella lista l")
# ritorna l'indice della prima occorrenza
index = l.index(9)
print("9 compare la prima volta in posizione", index, "nella lista l")
```

```python
""" Memorizza in un array di interi i voti ottenuti da sei studenti e determina il maggiore, il minore e la media"""
voti = []
for i in range(6):
    print("Voto studente ", i+1, ": ", end="")
    voti.append(int(input()))
    
# calcola maggiore, minore e media
max = voti[0]
min = voti[0]
media = 0
for v in voti:
    if v > max :
        max = v
    if v < min :
        min = v
    media = media + v

media = media / 6

print(voti)
print("Maggiore = ", max)
print("Minore = ", min)
print("Media = ", media)
```

```python
# per indicare i voti di tutti gli studenti
for i in range(len(voti)) :
    print("Voto studente ", i+1, " = ", voti[i])
```

### Numpy

```python
# Caricamento di una matrice

import numpy as np

# Richiesta delle dimensioni
numrow = int(input("Numero di righe = "))
numcol = int(input("Numero di colonne = "))

matr = np.zeros((numrow, numcol))

print("Caricamento della matrice")
for i in range(numrow) :
    for j in range(numcol) :
        print("inserisci elemnto di riga ", i, " e di colonna ", j,":", end = "" )
        matr[i,j] = int(input())
                        
# visualizza la matrice
for i in range(numrow):
    for j in range(numcol):
        print(matr[i,j]," ", end = "")
    print()
```

### Creazione di una classe di alunni

```python
# definiamo le nostre chiavi
chiavi_alunno = ("Nome", "Cognome", "Anno di corso", "Sezione", "Voti", "Assenze")

def inserisci_alunno(i):
    # crea un alunno con tutti i valori impostati a None
    alunno = dict.fromkeys(chiavi_alunno)
	# ora la funzione ci chiede di inserire i dati
    print()
    print("Inserire i dati dell'alunno ", i)
    for st in chiavi_alunno[:-2]:
        alunno[st] = input(st + ": ")
    return alunno

classe = []
numAl = int(input("Numero di alunni nella classe: "))
for i in range(numAl):
    classe.append(inserisci_alunno(i))
    

```

```python
classe
```

```python
chiavi_Materie = ("Italiano", "Matematica", "Latino", "Inglese")

def inserisci_voti_alunno(i):
    # crea un alunno con tutti i valori impostati a None
    valutazione = dict.fromkeys(chiavi_Materie)
	# ora la funzione ci chiede di inserire i dati
    print()
    print("Inserire i dati dell'alunno ", i)
    for st in chiavi_Materie:
        valutazione[st] = input(st + ": ")
    return valutazione

for i in range(numAl):
    classe[i]["Voti"] = inserisci_voti_alunno(i)
    
classe
```
