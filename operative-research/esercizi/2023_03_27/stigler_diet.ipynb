### Definisce una lista di nutrienti FoodList

```python
foods = pd.read_excel(xlsx, "Foods")
foodList = foods.Foods
```

# Importa la libreria GRB inizializza il problema 

```python
import gurobipy as gb
from gurobipy import GRB

model = gb.Model("Dieta")
```

# Definisce le variabili del problema (usando FoodList)

```python
xvar = model.addVars(foodList, name="x")
```

## Funzione Obiettivo
### Minimizzare la somma dei costi degli alimenti $$ \min  \sum_{f \in F} cost_f \ x_f $$
#### I dati nutrizionali riportati nel file excel si intendono per dollaro, non per unità, quindi l'obiettivo è determinare quanti dollari spendere per ogni alimento. I coefficienti di costo della funzione obiettivo sono dunque tutti pari a 1.

```python
obj = gb.quicksum(xvar[f] for f in foodList)
    
model.setObjective(obj, GRB.MINIMIZE)
```

### Vincoli del problema

### 1) Quantità minima di calorie
$$ \sum_{f \in F} \#calorie_f \ x_f \geq Daily\_Recommended\_Intake_{calorie}$$

```python
cnstr = gb.quicksum(foodList["Calories"][f]*xvar[f] for f in foodList ) >= nutrientList["Daily_Recommended_Intake"][0]
mod.addConstr(cnstr, name="MinCal")
```
