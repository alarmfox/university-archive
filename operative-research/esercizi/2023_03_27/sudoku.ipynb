# Sudoku 

### Il <span style="color:blue">Sudoku </span> è un gioco di logica in cui si deve inserire una cifra da 1 a 9 in ogni cella di una matrice 9x9, suddivisa a sua volta in 9 sottomatrici (riquadri) 3x3. 



### 1)  In alcuni elementi della matrice sono già presenti cifre. 

### 2)   Occorre inserire una cifra per ogni cella vuota in modo tale che <span style="color:blue">ogni riga, colonna e sottomatrice contenga ciascuna cifra una sola volta<span>.

### 3)   Affinchè il gioco sia valido, è necessario che i numeri già presenti facciano si che esista una e una sola soluzione.

### <span style="color:red">Risolvere il gioco significa individuare l'unica soluzione ammissibile (problema di feasibility)<span style>
